```python
import pandas as pd
import folium
```

```python
# lat: -3.7933116
# log: -38.5896559
mapa = folium.Map(location=[-3.7933116, -38.5896559], zoom_start=11)
mapa
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/estacoesbike.csv",
                 delimiter=';')
df.head()
```

```python
latitude = df["latitude"].values
longitude = df["longitude"].values
nomes = df["nome"].values
```

```python
mapa_recife = folium.Map(location=[-8.0540715,-34.9048899], zoom_start=12)

# Adicionando pontos ao mapa (markers)
for lat, lon, nome in zip(latitude, longitude, nomes):
  folium.Marker(location=[lat, lon], popup=nome).add_to(mapa_recife)

mapa_recife
```
