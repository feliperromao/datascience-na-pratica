```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

x = np.arange(10)
y = np.random.normal(size=10)

plt.plot(x, y)
plt.title("Título do gráfico")
plt.xlabel("Valores em X")
plt.ylabel("Valores em Y")
plt.show()
```

```python
x = np.arange(10)
y = np.random.normal(size=10)

fig, ax = plt.subplots();

ax.plot(x, y)
ax.set_title("Meu Titulo")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.tight_layout();
```

## Gerando dois graficos em uma unica figura

```python
np.random.seed(10)

x1 = np.arange(10)
y1 = np.random.normal(size=10)

x2 = np.arange(10)
y2 = 1 / np.random.normal(size=10)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,3))

# Gráfico 1
ax[0].plot(x1, y1)
ax[0].set_title("Gráfico 1")
ax[0].set_xlabel("X")
ax[0].set_ylabel("Y")

# Gráfico 2
ax[1].plot(x2, y2)
ax[1].set_title("Gráfico 2")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Y")

plt.tight_layout();
```
